```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

display(df['senioridade'].value_counts())



# df_senioridade = df['senioridade'].replace({
#     'SE': 'Senior',
#     'MI': 'Pleno',
#     'EN': 'Junior',
#     'EX': 'Executivo'
# })

# display(df_senioridade.value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

display(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})

display(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

display(df['remoto'].value_counts())
```

```python
# Teste usando as bibliotecas import matplotlib.pyplot as plt import seaborn as sns para vizualização dos dados


import matplotlib.pyplot as plt
import seaborn as sns


sns.histplot(df['salario'])
plt.show()


# sns.scatterplot(x='coluna_x', y='coluna_y', data=df)
# plt.show()
```

```python


sns.scatterplot(x='salario', y='salario', data=df)
plt.title('Relação entre Salário Original e Salário em USD')
plt.xlabel('Salário Original')
plt.ylabel('Salário em USD')
plt.show()
```

```python
df_clean = df.dropna()
df_clean['salario'] = df_clean['salario'].astype(float)

# Gráfico de distribuição de salários
plt.figure(figsize=(10,6))
sns.histplot(df_clean['salario'], bins=30, kde=True)
plt.title("Distribuição de Salários em USD")
plt.xlabel("Salário (USD)")
plt.ylabel("Quantidade")
plt.show()

# Gráfico de barras: média salarial dos 5 cargos mais comuns
top_cargos = df_clean['cargo'].value_counts().head(5).index
media_por_cargo = df_clean.groupby('cargo')['salario'].mean().loc[top_cargos]

plt.figure(figsize=(10,6))
media_por_cargo.plot(kind='bar')
plt.title("Média Salarial dos 5 Cargos Mais Comuns")
plt.xlabel("Cargo")
plt.ylabel("Média Salarial (USD)")
plt.show()
```
