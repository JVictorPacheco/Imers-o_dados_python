```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

display(df['contrato'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

display(df['remoto'].value_counts())
```

```python
import plotly.express as px


senioridade_media_salarial = df_clean.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salarial, x='senioridade', y='usd', title='Média Salarial por Senioridade', labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial Anual {USD}'})
fig.show()

```

```python
remoto_contagem = df_clean['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['Tipo de trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='Tipo de trabalho',
             values='quantidade',
             title='Distribuição de Remoto',
             hole=0.5
             )

fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()

# fig = px.bar(
#     senioridade_media_salarial,
#     x='senioridade',
#     y='usd',
#     title='Média Salarial por Senioridade', labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média Salarial Anual {USD}'})
# fig.show()

```

```python
# Análise das colunas e tipos de dados
print("📊 ESTRUTURA DETALHADA DOS DADOS")
print("="*50)

for col in df.columns:
    print(f"\n🔹 {col.upper()}:")
    print(f"   Tipo: {df[col].dtype}")
    print(f"   Valores únicos: {df[col].nunique()}")
    print(f"   Valores nulos: {df[col].isnull().sum()}")

    # Mostra alguns valores únicos para variáveis categóricas
    if df[col].dtype == 'object' and df[col].nunique() < 20:
        print(f"   Valores: {list(df[col].unique()[:10])}")
    elif df[col].dtype in ['int64', 'float64']:
        print(f"   Min: {df[col].min():.2f}, Max: {df[col].max():.2f}")
```
